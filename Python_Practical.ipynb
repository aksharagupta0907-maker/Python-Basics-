6. Write a function to demonstrate the behavior of mutable and immutable arguments

```python
def update_list(items):
    items.append(4)
    print(f"Inside function (mutable): {items}")

def update_number(number):
    number += 1
    print(f"Inside function (immutable): {number}")
```

```python
my_items = [1, 2, 3]
update_list(my_items)
print(f"Outside function (mutable): {my_items}")

my_number = 10
update_number(my_number)
print(f"Outside function (immutable): {my_number}")
```

9. Write a Python program to convert user input from string to integer, float, and boolean types


```python
user_input = input("Enter a value: ")
```

```python
try:
    int_val = int(user_input)
    print(f"As integer: {int_val}, Type: {type(int_val)}")
except ValueError:
    print("Cannot convert to integer")

try:
    float_val = float(user_input)
    print(f"As float: {float_val}, Type: {type(float_val)}")
except ValueError:
    print("Cannot convert to float")

bool_val = user_input.lower() == 'true'
print(f"As boolean: {bool_val}, Type: {type(bool_val)}")
```

13. Write a Python program to find the sum of all even numbers between 1 and 50

14. Write a program to reverse a string using a while loop

```python
input_string = input("Enter a string to reverse: ")
reversed_string = ""
length = len(input_string) - 1
```

```python
while length >= 0:
    reversed_string += input_string[length]
    length -= 1
```

```python
print(f"Reversed string: {reversed_string}")
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

```python
number = int(input("Enter a non-negative integer: "))
factorial = 1
i = 1
```

```python
if number < 0:
    print("Factorial does not exist for negative numbers.")
elif number == 0:
    print("The factorial of 0 is 1.")
else:
    while i <= number:
        factorial *= i
        i += 1
    print(f"The factorial of {number} is {factorial}")
```
