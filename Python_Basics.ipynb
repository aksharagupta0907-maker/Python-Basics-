7. What is the difference between “==” and “is” operators in Python?
 - The == operator checks for value equality (if two objects have the same value), while the is operator checks for identity equality (if two objects are actually the exact same object in memory).

10. What is the difference between implicit and explicit type casting?
  - Implicit: Python automatically converts one type to anther (e.g., adding an int to a float results in a float to avoid data loss).
  - Explicit: The programmer manually converts the type using functions like int(), float(), or str().

13. What is the difference between for and while loops?
 - For loop : Used for iterating over a sequence (like a list or range). You usually know how many times it will run,
 - while loops : Runs as long as a condition remains true. You might not know exactly how many iterations it will take.

14. Describe a scenario where a while loop is more suitable than a for loop?
 - A while loop is better when the number of iterations depends on a dynamic condition, such as:
